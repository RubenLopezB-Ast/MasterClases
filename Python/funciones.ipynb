```python
def saludar_nombre (nombre):  # Parámetro nombre que recibe por argumento -> "Ander" - "Victor" - "Daniel"
    print(f"Hola {nombre}!")

saludar_nombre("Ander")
saludar_nombre("Victor")
saludar_nombre("Daniel")
```

```python
def sumar_dos_numeros(numA, numB): #Los argumentos se pasan en el mismo orden en el que estén declarados los parámetros
    total = numA + numB
    print(total)

sumar_dos_numeros(5,8)
```

```python
def multiplicar_por_dos(numero):
    total = numero * 2
    return total #Declara el fin del función y expulsa el valor hacia afuera

multiplicar_por_dos(25)

```

```python
def saludar_usuario(nombre_usuario = "Paco"):
    print(f"Hola {nombre_usuario}")

saludar_usuario("Rubén")
saludar_usuario()
```

```python
x = 10 #Variable global está fuera de cualquier bloque de código

def mostrar_valor():
    x=5
    print("Valor dentro de la función", x) #Variable local está dentro de la función

mostrar_valor()

print("Valor fuera de la función", x, " es la global. ")
```

```python
# Documentación de funciones
def sumar_numeros(numA, numB):
    """
    Suma dos números y devuelve el resultado

    Args:
        numA (int): Primer número de la suma
        numB (int): Segundo número de la suma

    Returns:
        int: El resultado de la suma de ambos números
    """
    return numA + numB

sumar_numeros(5,7)
```
